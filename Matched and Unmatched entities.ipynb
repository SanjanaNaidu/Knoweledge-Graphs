```python
from rdflib import Graph, Literal, RDF, RDFS, Namespace
import spacy

# Load the RDF files into separate graphs
g1 = Graph()
g1.parse('learning_concepts (1).rdf')

g2 = Graph()
g2.parse('learning_concepts (2).rdf')

# Define the namespaces for the RDF files
ns1 = Namespace('http://example.com/ns1/')
ns2 = Namespace('http://example.com/ns2/')

# Extract the named entities from each graph
#named_entities1 = set(g1.objects(predicate=RDF.type, object=ns1.Entity))
#named_entities2 = set(g2.objects(predicate=RDF.type, object=ns2.Entity))
#Extract the named entities from each graph
named_entities1 = set(g1.objects(predicate=RDFS.label))
named_entities2 = set(g2.objects(predicate=RDFS.label))


# Exact matching function
def exact_matching(entities1, entities2):
    matched = []
    unmatched = []
    for e1 in entities1:
        if e1 in entities2:
            matched.append(e1)
        else:
            unmatched.append(e1)
    for e2 in entities2:
        if e2 not in entities1:
            unmatched.append(e2)
    return matched, unmatched

# Inexact matching function
def inexact_matching(entities1, entities2):
    matched = []
    unmatched = []
    for e1 in entities1:
        for e2 in entities2:
            if str(e1).lower() == str(e2).lower():
                matched.append(e1)
                entities2.remove(e2)
                break
        else:
            unmatched.append(e1)
    unmatched.extend(entities2)
    return matched, unmatched

# AI-based semantic matching function using spaCy
nlp = spacy.load("en_core_web_lg")
def semantic_matching(entities1, entities2):
    matched = []
    unmatched = []
    for e1 in entities1:
        doc1 = nlp(str(e1))
        max_sim = 0
        max_sim_entity = None
        for e2 in entities2:
            doc2 = nlp(str(e2))
            sim = doc1.similarity(doc2)
            if sim > max_sim:
                max_sim = sim
                max_sim_entity = e2
        if max_sim > 0.8:
            matched.append((e1, max_sim_entity))
            entities2.remove(max_sim_entity)
        else:
            unmatched.append(e1)
    unmatched.extend(entities2)
    return matched, unmatched

# Example usage
if __name__ == '__main__':
    entities1 = named_entities1
    entities2 = named_entities2

    # Exact matching
    exact_matched, exact_unmatched = exact_matching(entities1, entities2)
    print(f"Exact matched entities: {exact_matched}")
    print(f"Exact unmatched entities: {exact_unmatched}")

    # Inexact matching
    inexact_matched, inexact_unmatched = inexact_matching(entities1, entities2)
    print(f"Inexact matched entities: {inexact_matched}")
    print(f"Inexact unmatched entities: {inexact_unmatched}")

    # Semantic matching
    semantic_matched, semantic_unmatched = semantic_matching(entities1, entities2)
    print(f"Semantic matched entities: {semantic_matched}")
    print(f"Semantic unmatched entities: {semantic_unmatched}")
```
