To extract entity classes and named entities from an RDF file, we need to parse the RDF data using an RDF parsing library, such as RDFLib for Python, and then use RDF querying language, such as SPARQL, to extract the relevant information.

```python
from rdflib import Graph, Namespace

# Create an RDF graph and parse the RDF file
g = Graph()

g.parse("learning_concepts (2).rdf")

# Define the namespaces used in the RDF file
ns = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")

# Define the SPARQL query to extract the entity classes and named entities
query = """
    SELECT ?entity ?class ?label
WHERE {
  ?entity rdf:type ?class .
  FILTER (?class = ns:Person || ?class = ns:Place || ?class = ns:Organisation || ?class = ns:Event) .
  OPTIONAL { ?entity rdfs:label ?label }
  FILTER(lang(?label)='en')
}
"""

# Execute the SPARQL query on the RDF graph
results = g.query(query)

# Print the results
for row in results:
    entity = row.entity.split("/")[-1]
    class_name = row.class_name.split("/")[-1]
    print(f"Entity: {entity}, Class: {class_name}")
```

```python
from rdflib import Graph, URIRef

# Define the namespaces used in the RDF files
namespaces = {
    "dc": URIRef("http://purl.org/dc/elements/1.1/"),
    "foaf": URIRef("http://xmlns.com/foaf/0.1/"),
    "dbo": URIRef("http://dbpedia.org/ontology/"),
    "rdf": URIRef("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
}

# Define a function to extract entity classes and named entities from an RDF file
def extract_entities_from_rdf_file(rdf_file_path):
    # Create a new RDF graph and parse the RDF file into it
    rdf_graph = Graph()
    rdf_graph.parse(C:/Users/sanja/Downloads/learning_concepts (2).rdf, format="xml")

    # Define a dictionary to store the entities found in the RDF file
    entities = {}

    # Iterate over all of the subjects in the RDF graph
    for subject in rdf_graph.subjects():
        # Check if the subject is a named entity (has an "rdf:type" property)
        if (subject, namespaces["rdf"]+"type", None) in rdf_graph:
            # Get the type of the entity (the object of the "rdf:type" property)
            entity_type = rdf_graph.value(subject, namespaces["rdf"]+"type")

            # Check if the entity type is a known class
            if entity_type in [namespaces["dbo"]+"Person", namespaces["foaf"]+"Person"]:
                # Get the name of the person (the object of the "foaf:name" property)
                name = rdf_graph.value(subject, namespaces["foaf"]+"name")

                # Check if the name is a string (not a URI reference)
                if isinstance(name, str):
                    # Add the person to the dictionary of entities
                    entities[str(subject)] = {
                        "type": str(entity_type),
                        "name": name
                    }

    # Return the dictionary of entities found in the RDF file
    return entities

# Call the function on each of the RDF files to extract entities from them
rdf_file_paths = ["learning_concepts (2).rdf"]
for rdf_file_path in rdf_file_paths:
    entities = extract_entities_from_rdf_file(rdf_file_path)
    print(f"Entities found in {rdf_file_path}:")
    print(entities)
```
